```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('gots_test_1000_entities.csv')
df.head()
```

# Which countries have the most certified companies?

```python
country_counts = df['country'].value_counts()
print(country_counts.head())
```

```python
# Plot a bar chart
plt.figure(figsize=(20, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='plasma')
plt.xlabel('Country')
plt.ylabel('Certified Companies')
plt.title('Number of Certified Companies by Country')
plt.xticks(rotation= 45, ha='right')
plt.tight_layout()
plt.show()
```

# What are the most certified product categories?


```python
categories_df = df['product_category'].str.split(',').explode().str.strip()
product_category_counts = categories_df.value_counts()
print(product_category_counts.head())
```

```python
# Plot a bar chart
plt.figure(figsize=(12, 5))
sns.barplot(x=product_category_counts.index, y=product_category_counts.values, palette='plasma')
plt.xlabel('Product Category')
plt.ylabel('Certified Companies')
plt.title('Number of Certified Companies by Product Category')
# plt.xticks(rotation=45)
plt.xticks(rotation= 45, ha='right')
plt.tight_layout()
plt.show()
```

# Identify the most common Certification Body.

```python
certification_body_counts = df['certification_body'].value_counts()
print(certification_body_counts)
```

```python

common_certification_body = certification_body_counts.idxmax()
print(common_certification_body)
```

```python
# Plot a bar chart
plt.figure(figsize=(12, 3))
sns.barplot(x=certification_body_counts.index, y=certification_body_counts.values, palette='plasma')
plt.xlabel('Certification Body')
plt.ylabel('Certified Companies')
plt.title('Number of Certified Companies by Certification Body')
plt.xticks(rotation= 45, ha='right')
plt.tight_layout()
plt.show()
```
